#### 1. Transformar as colunas numericas para categóricas

 Utilizando os dicionários de mapeamento fornecidos. Isso tornará a análise mais intuitiva e facilitará a interpretação dos resultados.


```python
# Dicionários de mapeamento
map_turno = {"1": "Administrativo", "2": "Turno rotativo"}
map_genero = {"1": "Homem", "2": "Mulher", "3": "Pessoa não-binária", "4": "Prefiro não responder", "5": "Outro"}
map_cor = {"1": "Branca", "2": "Preta", "3": "Amarela", "4": "Parda", "5": "Indígena", "6": "Prefiro não responder"}
map_faixa = {"1": "Menos de 19 anos", "2": "19 a 24 anos", "3": "25 a 34 anos", "4": "35 a 44 anos", "5": "45 a 54 anos", "6": "55 a 64 anos", "7": "65 anos ou mais"}
map_escolaridade = {
    "1": "Ensino fundamental incompleto", "2": "Ensino fundamental completo", "3": "Ensino médio incompleto",
    "4": "Ensino médio completo", "5": "Ensino superior incompleto", "6": "Ensino superior completo", "7": "Pós graduação completa"
}
map_tempo_empresa = {
    "1": "Menos de 1 ano", "2": "De 1 a 3 anos", "3": "De 4 a 6 anos", "4": "De 7 a 10 anos", "5": "De 11 a 15 anos", "6": "Mais de 15 anos"
}
map_tempo_funcao = {
    "1": "Menos de 1 ano", "2": "De 1 a 3 anos", "3": "De 4 a 6 anos", "4": "De 7 a 10 anos", "5": "De 11 a 15 anos", "6": "Mais de 15 anos"
}

# Aplicar mapeamento nas colunas

# Função padrão de limpeza
def limpar_coluna(col):
    return (
        col.astype(str)
        .str.replace('.0', '', regex=False)
        .str.strip()
    )

# Aplicar mapeamento nas colunas
df['Turno'] = limpar_coluna(df['Turno']).replace(map_turno)
df['Gênero'] = limpar_coluna(df['Gênero']).replace(map_genero)
df['Cor/raça autodeclarada'] = limpar_coluna(df['Cor/raça autodeclarada']).replace(map_cor)
df['Faixa etária'] = limpar_coluna(df['Faixa etária']).replace(map_faixa)
df['Escolaridade'] = limpar_coluna(df['Escolaridade']).replace(map_escolaridade)
df['Tempo na empresa'] = limpar_coluna(df['Tempo na empresa']).replace(map_tempo_empresa)
df['Tempo na função atual dentro da empresa'] = limpar_coluna(df['Tempo na função atual dentro da empresa']).replace(map_tempo_funcao)
df.head()
```

#### Observação no metadados

Vou criar 2 colunas a partir da coluna 'Insira abaixo o código referente a seu cargo e área', para facilitar a análise de área e cargo. No final remover a coluna original.

```python
# CRIAR COLUNAS
# ==========================================
df['area'] = df['Insira abaixo o código referente a seu cargo e área']
df['cargo'] = df['Insira abaixo o código referente a seu cargo e área']

# DICIONÁRIOS DE MAPEAMENTO
# ==========================================
map_area = {
    "1": "Operação",
    "2": "Operação",
    "3": "Operação",
    "4": "Manutenção",
    "5": "Manutenção",
    "6": "Manutenção",
    "7": "Logística",
    "8": "Logística",
    "9": "Logística",
    "10": "RH",
    "11": "RH",
    "12": "RH",
    "13": "Segurança do Trabalho",
    "14": "Segurança do Trabalho",
    "15": "Segurança do Trabalho",
    "16": "Administrativo",
    "17": "Administrativo",
    "18": "Administrativo",
    "19": "Diretoria"
}

map_cargo = {
    "1": "Operador(a) de Produção",
    "2": "Líder de Produção",
    "3": "Supervisor(a) Operacional",
    "4": "Técnico(a) de Manutenção",
    "5": "Planejador(a) de Manutenção",
    "6": "Supervisor(a) de Manutenção",
    "7": "Assistente de Logística",
    "8": "Analista de Logística",
    "9": "Coordenador(a) de Logística",
    "10": "Assistente de RH",
    "11": "Analista de RH",
    "12": "Coordenador(a) de RH",
    "13": "Técnico(a) de Segurança do Trabalho",
    "14": "Engenheiro(a) de Segurança",
    "15": "Coordenador(a) de Segurança",
    "16": "Assistente Administrativo(a)",
    "17": "Analista Administrativo(a)",
    "18": "Coordenador(a) Administrativo(a)",
    "19": "Gerente Geral"
}

# limpar as colunas
df['area'] = (df['area'].astype(str).str.replace('.0', '', regex=False).str.strip())
df['cargo'] = (df['cargo'].astype(str).str.replace('.0', '', regex=False).str.strip())

# aplicar o mapeamento
df['area'] = df['area'].replace(map_area)
df['cargo'] = df['cargo'].replace(map_cargo)

# remover a coluna original
df = df.drop(columns=['Insira abaixo o código referente a seu cargo e área'])

# ver o resultado. 
df[['area', 'cargo']].head()
```

418 usuarios responderam a pesquisa de forma completa e 55 ainda não completou. Com isso, vou fazer um filtro para analise de apenas os dados completos.

```python
# Filtrar apenas respostas completas
df_completed = df.query('`Status de resposta` == "Completed"')
df_completed.head()
```

```python
# Verificar se ainda existem valores nulos na base filtrada
print(df_completed.isnull().values.any())
```

```python
# Reorganizar colunas para colocar as categóricas no início e as numéricas no final

#  Seleciona colunas categóricas
cat_cols = df_completed.select_dtypes(include=['object', 'category']).columns

# Seleciona demais colunas
num_cols = df_completed.select_dtypes(exclude=['object', 'category']).columns

# Reorganiza o dataframe
df_completed = df_completed[list(cat_cols) + list(num_cols)]

# Coloca 'ID de resposta' como primeira coluna
cols = ['ID de resposta'] + [
    col for col in df_completed.columns
    if col != 'ID de resposta'
]

df_completed = df_completed[cols]
```
